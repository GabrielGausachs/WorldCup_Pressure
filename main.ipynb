#### Data Loading and Preprocessing

````python
# Load passes and tracking_passes data
passes_df, tracking_passes_df = load_passes(BASE_PATH)

# Map player positions
passes_df = map_positions(BASE_PATH, passes_df)

# Clean missing rows in passes_df and tracking_passes_df
passes_df, tracking_passes_df = clean_missing_rows(passes_df, tracking_passes_df)

# Add minutes context
passes_df = add_minutes_context(BASE_PATH, passes_df)
````
